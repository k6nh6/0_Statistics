```python
import numpy as np
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
speed.sort()
x = np.mean(speed)
print(x)

y=np.median(speed)
print(y)
```

```python
from scipy import stats
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
z = stats.mode(speed, keepdims=False)
print(z)
```

```python
# Standard Deviation
import numpy

speed = [86,87,88,86,87,85,86]

x = numpy.std(speed)

print(x) 
```

```python
# Variance 
import numpy
import math
speed = [32,111,138,28,59,77,97]

x = numpy.var(speed)

print(math.sqrt(x))
```

<h1>Linear Regression
    <h2>The term regression is used when you try to find the relationship between variables.

In Machine Learning, and in statistical modeling, that relationship is used to predict the outcome of future events.

```python
#Example

#Import scipy and draw the line of Linear Regression:
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#slope of line,the y intercept of line, r the correlation cofficent, p-value, accuracy of slope estimate
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show() 
```
