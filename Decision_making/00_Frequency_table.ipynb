```python
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
x = [2, 3, 4, 5,6,7] #number of memebers in a family 
frequency = [3, 7, 10, 5, 4,1] #occurences of family with above family number

# Calculate cumulative and reverse cumulative frequencies
cumulative_frequency = [sum(frequency[:i + 1]) for i in range(len(frequency))]
reverse_cumulative_frequency = [sum(frequency[i:]) for i in range(len(frequency))]

# Plot the graph
plt.plot(x, cumulative_frequency, marker="^", label='Cumulative Frequency')
plt.plot(x, reverse_cumulative_frequency, label='Reverse Cumulative Frequency',marker="s")

# Add horizontal lines starting from each point on the y-axis
for y_value in [5,10,15,20,25,30,35]:
    plt.axhline(y=y_value, color='gray',  alpha=0.3)

plt.title("Pictorial representation of frequencies")
plt.xlabel('Number')
plt.ylabel('Cummulative Frequencies')
plt.legend(loc='lower left', bbox_to_anchor=(-0.3, -0.3))
plt.show()
```
